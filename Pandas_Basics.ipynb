##Pandas

```python
#Importing Pandas
import pandas as pd
```

##Pandas Series
* 1D Array
* like column in Table
* Default Index / Row Name

```python
l=[1,2,3,4,5]
s=pd.Series(l)
print(s)
print(type(s))
```

```python
lst=["Awais","Bob","Charlie","Ahmad"]
ser=pd.Series(lst)
print(ser)
```

```python
lst1=["Awais","Bob",12,"Ahmad"]
ser1=pd.Series(lst1)
print(ser1)
```

```python
#Setting Index
lst=["Awais","Bob","Charlie","Ahmad"]
ser=pd.Series(lst,index=["1st","2nd","3rd","4th"])
print(ser)
```

```python
#Series Operation
print(s+2)
print("\n",s**2)
```

```python
#Series From 1D Array
import numpy as np
arr=np.array([1,2,3,4,5])
ser=pd.Series(arr)
print(ser)
```

```python
#Series From Dictonary
dic={"a":[1,2],"b":(3,4),"c":5}
ser=pd.Series(dic)
#Keys will be Considered as Index name
print(ser)
```

```python
print(ser.ndim)
print(ser.shape)
print(ser.size)
```

##Pandas DataFrame
* 2 Dimensional Data Structure
* Rows and Columns(Tabular)

```python
dic={"Name":["Awais","Ahmad","Ali"],"Age":[22,23,24]}
df=pd.DataFrame(dic) #Keys as Col Label
print(df)
```

```python
l=np.array([[1,2,3,4],[11,22,33,44]])
arrdf=pd.DataFrame(l)
print(arrdf)
```

```python
#Setting Column and Row Labels(Names)
df2=pd.DataFrame(l,columns=["A","B","C","D"],index=["A","B"])
print(df2)
```

```python
#Getting Column and Row Names
print(df2.index)
print(df2.columns)
```
