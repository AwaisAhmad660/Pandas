Working with missing Values

```python
#Creating copy orignal data
df_copy=df.copy()
```

```python
df_copy["age"].isnull().sum()
```

```python
df_copy["age"].fillna(0)
df_copy["age"].fillna(df_copy["age"].mean())

```

```python
df_copy["age"].isnull().sum()
```

```python
#ffill->forward fill (fill with previous cell data )
df_copy["age"].ffill()
# df_copy
```

```python
#bfill->forward fill (fill with after cell data )
df_copy["age"].bfill()
```

```python
#Remove empty values

df_copy.dropna()
```

```python
#droping column wise
df_copy.dropna(axis=1)
```

**Duplicate Values**

```python
data = {
    "order_id": [1001, 1002, 1003, 1004, 1005, 1001, 1003, 1006, 1006],
    "customer": ["Ali", "Ayesha", "Ahmed", "Sara", "Ali", "Ali", "Ahmed", "Hassan", "Hassan"],
    "city": ["Lahore", "Karachi", "Lahore", "Islamabad", "Lahore",
             "Lahore", "Lahore", "Karachi", "Karachi"],
    "category": ["Electronics", "Electronics", "Furniture", "Electronics",
                 "Electronics", "Electronics", "Furniture",
                 "Accessories", "Accessories"],
    "quantity": [1, 2, 1, 3, 1, 1, 1, 2, 2],
    "price": [50000, 60000, 30000, 45000, 50000, 50000, 30000, 15000, 15000]
}

df = pd.DataFrame(data)
df
```

```python
#detecting Duplicate values
df[df.duplicated()]
```

```python
#droping dubplicate Values
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.drop_duplicates().reset_index() #reseting indexes
```

```python
df["order_id"].value_counts()
```
