```python
import pandas as pd
```

**Groupby :**
group data based on one or more columns and then apply an aggregation (sum, mean, count, etc.).
* group columns that are in text formate or catagorical data

```python
data = {
    "order_id": [1001, 1002, 1003, 1004, 1005, 1001, 1003, 1006, 1006],
    "customer": ["Ali", "Ayesha", "Ahmed", "Sara", "Ali", "Ali", "Ahmed", "Hassan", "Hassan"],
    "city": ["Lahore", "Karachi", "Lahore", "Islamabad", "Lahore",
             "Lahore", "Lahore", "Karachi", "Karachi"],
    "category": ["Electronics", "Electronics", "Furniture", "Electronics",
                 "Electronics", "Electronics", "Furniture",
                 "Accessories", "Accessories"],
    "quantity": [1, 2, 1, 3, 1, 1, 1, 2, 2],
    "price": [50000, 60000, 30000, 45000, 50000, 50000, 30000, 15000, 15000]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby("category")["price"].sum()
```

```python
df.groupby("city")["price"].sum()
```

```python
df.groupby("city")["price"].count()
```

```python
df.groupby("city")[["price","customer"]].min()
```

```python
df.groupby("category")[["price","quantity"]].sum()
```
